## Beat Frequency ##
Demonstrate generation and output of audio data using numpy, Jupyter and matplotlib

Taken from http://ipywidgets.readthedocs.io/en/latest/examples/Beat%20Frequencies.html 

```python
# -*- coding: utf-8 -*-
from __future__ import division, print_function

import numpy as np

import matplotlib.pyplot as plt

from ipywidgets import interactive
from IPython.display import Audio, display
```

```python
def beat_freq(f1 = 220.0, f2 = 224.0):
    """
    Overlay two sine functions with frequencies f1 and f2 and create an intereactive audio player.
    """
    max_time = 3
    rate = 8000
    t = np.linspace(0, max_time, rate*max_time)
    s = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2*t)
    display(Audio(data=s, rate=rate));
    return s
```

Add two sine signals with slighty different frequencies and watch / listen to the beat frequency:

```python
v = interactive(beat_freq, f1=(200.0, 300.0), f2=(200.0, 300.0));
display(v); # needed to generate the data?!
plt.plot(v.result[0:6000]);
```

The rest of the code demonstrates some options for interacting with object *v* 

```python
v.kwargs # return the keyword arguments of v
```

```python
f1, f2 = v.children[:2] # access and 
f1.value = 255          # override the
f2.value = 260          # slider settings
plt.plot(v.result[0:6000]);
```
